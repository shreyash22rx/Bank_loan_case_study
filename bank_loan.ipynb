# Univariate Analysis

```python
Numarical_Data = ['AMT_ANNUITY','AMT_GOODS_PRICE','CNT_FAM_MEMBERS']
```

```python
Categorical_Data = ['FLAG_OWN_CAR','FLAG_OWN_REALTY','NAME_TYPE_SUITE','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS','NAME_HOUSING_TYPE','OCCUPATION_TYPE',]
```

```python
def Uni_Analysis_Numarical(dataframe, column):
  sns.set(style='darkgrid')
  plt.figure(figsize=(25, 5))
  plt.subplot(1, 3, 1)
  sns.boxplot(data=dataframe, x=column, orient='v').set(title='Box Plot')
  plt.subplot(1, 3, 2)
  sns.distplot(dataframe[column].dropna()).set(title='Distplot')
  plt.show()
```

```python
def Uni_Analysis_Categorcal(dataframe, column):
  sns.set(style='darkgrid')
  plt.figure(figsize = [12,5])
  dataframe[column].value_counts().plot.barh(width = 0.8)
  plt.title(column)
  plt.show()
```

```python
for i in Numarical_Data:
  Uni_Analysis_Numarical(pd_application,i)
```

```python
for i in Categorical_Data:
  Uni_Analysis_Categorcal(pd_application,i)
```
